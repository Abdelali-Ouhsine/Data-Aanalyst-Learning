```python
df["is_weekend"] = df["is_weekend"].astype(int)
df["is_holiday"] = df["is_holiday"].astype(int)
df["season"] = df["season"].astype(int)
df["weather_code"] = df["weather_code"].astype(int)
```

```python
df = df.rename(columns={"cnt":"Nombre de trajets"})
df = df.rename(columns={"t1":"Température réelle (°C)"})
df = df.rename(columns={"t2":"Température ressentie (°C)"})
df = df.rename(columns={"hum":"Humidité"})
df = df.rename(columns={"wind_speed":"Vitesse du vent (km/h"})
df = df.rename(columns={"weather_code":"Météo"})
df = df.rename(columns={"season":"Saison"})
```

```python
df['Humidité'] = df['Humidité'] / 100
```

```python
Saison_map = {
    0 : 'Printemps', 1 : 'Été', 2 : 'Automne', 3 : 'Hiver'
}
df["Saison"] = df["Saison"].map(Saison_map)
df
```

```python
Météo_map = {
    1 : 'Clair', 2 : 'Nuages épars', 3 :' Nuages fragmentés', 4 : 'Couvert,', 7 : 'Pluie', 10 : 'Pluie avec orage' , 26 : 'Neige'
}
df["Météo"] = df["Météo"].map(Météo_map)
df.head
```

```python
#la humidite doit etre entre 0 et 1
assert df["Humidité"].between(0,1).all()
```

```python
df.to_csv("london_road.xlsx")
```
